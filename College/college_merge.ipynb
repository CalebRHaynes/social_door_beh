```python
import glob 
import pandas as pd
from functools import reduce
```

```python
for i in glob.glob('behavioral/s*'):
    #print(i[-4:])
    
    #doors
    sub_door_file_list = glob.glob(i + '/doors/onset/*')
    df_r1_list = []
    df_r2_list = []
    for j in sub_door_file_list:
        
        trial_type = j[29:-4]
        
        import csv

        with open(j, newline='') as f:
            reader = csv.reader(f)
            h = list(reader)
        flat_list = [item for sublist in h for item in sublist]
        run1_onsets = flat_list[0].split()
        run1_onsets_float = [float(i) for i in run1_onsets]
        
        run2_onsets = flat_list[1].split()
        run2_onsets_float = [float(i) for i in run2_onsets]
                
        
        trial_dict = {'onset': run1_onsets_float}
        trial_dict2 = {'onset': run2_onsets_float}
        
        
        r1 = pd.DataFrame(trial_dict)
        r1['trial_type'] = trial_type[:-6]
        
        r2 = pd.DataFrame(trial_dict2)
        r2['trial_type'] = trial_type[:-6]
        
        
        df_r1_list.append(r1)
        df_r2_list.append(r2)
        
        
        #df_r1['onset'].append(run1_onsets)
        
        #run2_onsets = flat_list[1].split()
        #print(run2_onsets)
        #h['trial_type'] = j[29:-4]
    
    df_r1 = pd.concat(df_r1_list)
    df_r1 = df_r1.sort_values(by = 'onset')
    
    df_r1 = df_r1[~df_r1['trial_type'].isin(['Instruction'])] 
    
    df_r2 = pd.concat(df_r2_list)
    df_r2 = df_r2.sort_values(by = 'onset')
    
    df_r2 = df_r2[~df_r2['trial_type'].isin(['Instruction'])] 
    
    
    #add durations
    def duration_label(row):
        if row['trial_type'] == "Lose_Instruction":
            return 5000
        if row['trial_type'] == "LoseGames_Ant":
            return 600
        if row['trial_type'] == "LoseGames_FB_Lose":
            return 1000
        if row['trial_type'] == "LoseGames_FB_Win":
            return 1000
        if row['trial_type'] == "Prize_Instruction":
            return 5000
        if row['trial_type'] == "WinGames_Ant":
            return 600
        if row['trial_type'] == "WinGames_FB_Lose":
            return 1000
        if row['trial_type'] == "WinGames_FB_Win":
            return 1000
        if row['trial_type'] == "LoseGames_ITI":
            return 1100
        if row['trial_type'] == "WinGames_ITI":
            return 1100
        
    

    df_r1['duration'] = df_r1.apply(lambda row: duration_label(row), axis=1)    
    df_r2['duration'] = df_r2.apply(lambda row: duration_label(row), axis=1)    
            
    
    
     
    df_r1.to_csv('sub-' + i[-4:] + '_doors_run-1.tsv', sep = "\t", index=False)
    df_r2.to_csv('sub-' + i[-4:] + '_doors_run-2.tsv', sep = "\t", index=False)
    

```
